# Models

## Scaling the data 

```python
#let's show the improvement that occurs when we scale our data 
#Normalisation 
mms=MinMaxScaler()
#Standardization
ss=StandardScaler()

data['Oldpeak']=mms.fit_transform(data[['Oldpeak']])
data['Age'] = ss.fit_transform(data[['Age']])
data['RestingBP'] = ss.fit_transform(data[['RestingBP']])
data['Cholesterol'] = ss.fit_transform(data[['Cholesterol']])
data['MaxHR'] = ss.fit_transform(data[['MaxHR']])
data.head()
```

### Selection of train and test datasets

```python
X = data.drop('HeartDisease' ,axis=1)
y= data['HeartDisease']
```

```python
#train_test split 
#what is the random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#dimensions of our dataframes
print("Train dimensions : " ,X_train.shape)
print("Test dimensions : ",X_test.shape)
```

## Model function

Première version de test de nos différents modèles. 

```python
#general function to test our different models 

def model(classifier,X_train,y_train,X_test,y_test):
    model=classifier
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    #cross validation
    cv = RepeatedStratifiedKFold(n_splits = 10,n_repeats = 3,random_state = 1)
    #Accuracy 
    print("Accuracy :",accuracy_score(y_test,y_pred))
    #Cross validation score
    print("Cross Validation Score : ",cross_val_score(classifier,X_train,y_train,cv = cv,scoring = 'roc_auc').mean())
	#ROC_AUC_Score
    print("ROC_AUC Score : ",'{0:.2%}'.format(roc_auc_score(y_test,y_pred)))
    RocCurveDisplay.from_estimator(classifier, X_test,y_test)
    print(classification_report(y_test,y_pred))
    
    
```

### Evaluer la qualité d'un modele 

- matrice de confusion
- precision
- recall
- f1-score

```python
def get_confusion_matrix(classifier,X_train,y_train,X_test,y_text):
    '''
    classifier : The classifier we want to evaluate 
    y_test: true labels
   	return : dataframe confusion matrix
    '''
    classifier.fit(X_train,y_train)
    cm = confusion_matrix(y_test, classifier.predict(X_test))
    cm_df = pd.DataFrame(cm)
    return cm_df

def plot_confusion_matrix(cm_df, title='Confusion Matrix', cmap=plt.cm.Blues):
    '''
    cm_df: confusion matrix dataframe
    title: title of the plot
    cmap: color map
    '''
    plt.figure(figsize=(8,6))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap=cmap)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def get_accuracy(cm_df):
    '''
    cm_df: confusion matrix dataframe
    '''
    total = cm_df.values.sum()
    correct = np.trace(cm_df.values)
    return correct / total

def get_precision(cm_df):
    '''
    cm_df: confusion matrix dataframe
    return [precision for 0 , precision for 1]
    '''
    col_sum = cm_df.sum(axis=0)
    precision = []
    for i in range(cm_df.shape[0]):
        precision.append(cm_df.iloc[i,i] / col_sum[i])
    return precision


def get_recall(cm_df):
    '''
    cm_df: confusion matrix dataframe
    '''
    row_sum = cm_df.sum(axis=1)
    recall = []
    for i in range(cm_df.shape[0]):
        recall.append(cm_df.iloc[i,i] / row_sum[i])
        
    return recall

def get_f1_score(cm_df):
    '''
    cm_df: confusion matrix dataframe
    '''
    precision = get_precision(cm_df)
    recall = get_recall(cm_df)
    f1_score = []
    for i in range(cm_df.shape[0]):
        f1_score.append(2 * precision[i] * recall[i] / (precision[i] + recall[i]))
    return f1_score
```

## Logistic regression


```python
#comprendre pourquoi avec une valeur d'iteration = 100 on avait 
#un problème de convergence 
lr_classifier=LogisticRegression(random_state = 0,C=10,penalty= 'l2',max_iter = 1000) 
```

```python
model(lr_classifier,X_train,y_train,X_test,y_test)
```

```python
#the confusion matrix of Logistic Regression 
cm_df_lr=get_confusion_matrix(lr_classifier,X_train,y_train,X_test,y_test)
plot_confusion_matrix(cm_df_lr)
```

```python
#evaluate the Logistic Regression model
print("Precision :", get_precision(cm_df_lr))
print("Recall :",get_recall(cm_df_lr))
print("F1-score :",get_f1_score(cm_df_lr))
```

Maintenant on souhaite effectuer un grid search pour avoir les meilleurs hyperparametres possibles et ainsi assurer les performances de notre regression logistique sachant que les hyperparametres de la regression logistique sont :
- penalty
- C
- Solver 
- max-iter 

```python
def LR_with_grid_search(classifier, X_train, y_train, X_test, y_test):
    # Define the parameter grid to search
    param_grid = {'C': [0.1, 1, 10, 100, 1000], 'penalty': ['l1', 'l2'], 'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
    
    # Create an instance of GridSearchCV
    grid = GridSearchCV(classifier, param_grid, cv = 5, scoring = 'roc_auc')
    
    # Fit the grid search to the training data
    grid.fit(X_train, y_train)
    
    # Print the best hyperparameters found by the grid search
    print("Best hyperparameters: ", grid.best_params_)
    
    # Use the best classifier found by the grid search
    classifier = grid.best_estimator_
    
    # Fit the classifier to the training data
    classifier.fit(X_train, y_train)
    
    # Predict the class labels for the test data
    y_pred = classifier.predict(X_test)
    
    # Cross validation
    cv = RepeatedStratifiedKFold(n_splits = 10, n_repeats = 3, random_state = 1)
    
    # Accuracy 
    print("Accuracy:", accuracy_score(y_test, y_pred))
    
    # Cross validation score
    print("Cross Validation Score:", cross_val_score(classifier, X_train, y_train, cv = cv, scoring = 'roc_auc').mean())
    
    # ROC-AUC Score
    print("ROC-AUC Score:", '{0:.2%}'.format(roc_auc_score(y_test, y_pred)))
    
    # Display the ROC curve
    RocCurveDisplay.from_estimator(classifier, X_test, y_test)
    
    # Print the classification report
    print(classification_report(y_test, y_pred))
```

```python
import warnings
warnings.filterwarnings("ignore")
LR_with_grid_search(lr_classifier,X_train,y_train,X_test,y_test)
```
